```python
import numpy as np
A = np.array([[2, -1],[-1, 2]]) # AX = B
B = np.array([0, 2])
x = np.linalg.solve(A, B)
print(x)
```

```python
A = np.array([[2, -1],[-1, 2]])
a = np.linalg.inv(A)
print( np.dot(a,B))
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.grid()
a = 0      # start point
A1x = [a, A[0][0]]
A1y = [a, A[1][0]]
plt.plot(A1x, A1y, label= 'VA1, the first column vector of A')  

SA1x = [i*x[0] for i in A1x]
SA1y = [i*x[0] for i in A1y]
plt.plot(SA1x, SA1y, label = 'VA1 scaled by x[0]')       # lines overlap

A2x = [a, A[0][1]]
A2y = [a, A[1][1]]
plt.plot(A2x, A2y, label = 'the second column vector of A')  
slope = (A[1][1]-a) / (A[0][1]-a)

dA2x = A[0][1]-a          # the scale of x axis of A2
dSA2x = x[1]*dA2x
x0 = SA1x[-1]
x1 = SA1x[-1]+dSA2x

SA2x = [x0, x1] 
y0 = SA1y[-1]
SA2y = []
for i in SA2x:  
    j = slope*(i-x0)+ y0       # determine y axis by linear expression 
    SA2y.append(j)   
plt.plot(SA2x, SA2y, label = 'VA1 scaled by x[1]')   

Bx = [SA1x[0], SA2x[-1]]
By = [SA1y[0], SA2y[-1]]           # from the start point of SA1 to the last point of SA2
plt.plot(Bx, By, label='Linear Combination of VA1 and VA2')
plt.legend()




 




```
